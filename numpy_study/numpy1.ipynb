# ndarray
## ndarray特性
### 多维性

```python
import numpy as np
```

```python
arr = np.array(5) #0维
print(arr)
print('维度',arr.ndim) #打印维度
```

```python
arr = np.array([1,2,3]) #1维
print(arr)
print('维度',arr.ndim) #打印维度
```

```python
arr = np.array([[1,2,3],[4,5,6]]) #2维
print(arr)
print('维度',arr.ndim) #打印维度
```

### 同质性

```python
arr = np.array([1,'hello'])  #不同的数据类型会自动转换
print(arr)
```

```python
arr = np.array([1,2.5])  #不同的数据类型会自动转换
print(arr)
print(arr.dtype)
```

## ndarray属性

```python
arr = np.array(1)
print(arr)
print('数组形状',arr.shape)
print('元素个数',arr.size)
print(arr.dtype)
print('元素转置',arr.T)
```

```python
arr = np.array([1,2,3])
print(arr)
print('数组形状',arr.shape)
print('元素个数',arr.size)
print(arr.dtype)
print('元素转置',arr.T)
```

```python
arr = np.array([[1,2.5,3],[3,4,5]])
print(arr)
print('数组形状',arr.shape)
print('元素个数',arr.size)
print(arr.dtype)
print('元素转置',arr.T)
```

## ndarray的创建

```python
# 基础创建方法
list1 = [4,5,6]
arr = np.array(list1,dtype=np.float64)
print(arr)
```

```python
# copy方法
arr1 = np.copy(arr) #只是元素相同
print(arr1)
arr1[0] = 2
print(arr)
```

```python
# 预定义形状
# 全0 全1 未初始化 固定值

#全0
arr = np.zeros((2,3))
print(arr)
print(arr.dtype)

arr1 = np.zeros((4,))
print(arr1)
```

```python
# 全1
arr = np.ones((2,3),dtype = np.int64)
print(arr)
```

```python
# 未定义
arr = np.empty((4,3))
print(arr)
```

```python
# 固定值
arr = np.full((3,3),2026,dtype=float)
print(arr)
arr1 = np.zeros_like(arr)
print(arr1)
```

```python
#  等差数列
arr = np.arange(2,10,2)  #起始，终止（不包含），步长
print(arr)
```

```python
# 等间隔数列
arr = np.linspace(0,10,5) #起始，终止，个数
print(arr)
arr1 = np. arange(0,11,2.5)
print(arr1)
```

```python
# 对数间隔数列
arr = np.logspace(0,4,3,base = 2)
print(arr)
```

```python
#单位矩阵
arr = np.eye(4,3,dtype=int)
print(arr)
```

```python
# 对角矩阵
arr = np. diag([1,2,3,4])
print(arr)
```

```python
# 随机数组生产
# 生成0-1之间的随机浮点数（均匀）
arr = np.random.rand(2,3)
print(arr)
# 指定范围之间
arr1 = np.random.uniform(3,6,(2,3))
print(arr1)

#整数
arr2 = np.random.randint(3,100,(2,3))
print(arr2)
```

```python
# 生产随机数列（正态分布）（-3到3）
#两边概率小，中间概率大
arr = np.random.randn(2,3)
print(arr)
```

```python
#种子
np.random.seed(20)
arr = np.random.randint(1,10,(2,5))
print(arr)
```
