# ndarray数据类型
布尔 bool

浮点 float
复数 complex


```python
import numpy as np
```

```python
arr = np.array([1,0,2,0],dtype=np.bool)
print(arr)
```

```python
arr = np.array([1,0,2,0],dtype=np.int64)
print(arr)
```

# 索引和切片

```python
# 一维数组
np.random.seed(1)
arr = np.random.randint(1,100,30)
print(arr)
print(arr[0])
print(arr[:]) #all
print(arr[2:5]) #左包右不包
```

```python
print(arr[(arr>10)&(arr<50)])
```

```python
#slice切片
print(arr[slice(2,15,3)])
```

```python
# 二维数组
np.random.seed(1)
arr = np.random.randint(1,100,(3,3))
print(arr)
print(arr[1][1])
print(arr[1,1:3])
print(arr[arr>50])
print(arr[2][arr[2]>50])
print(arr[:,2])
```
