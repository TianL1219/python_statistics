# 统计函数

```python
import numpy as np
```

```python
np.random.seed(2)
arr = np.random.randint(1,20,8)
print(arr)
print(np.sum(arr)) #求和
print(np.mean(arr)) #平均值
print(np.median(arr)) #中位数
```

```python
#标准差 方差
print(np.var([1,2,3]))
print(np.std([1,2,3]))
```

```python
print(np.max(arr))
print(np.min(arr))
print(np.percentile(arr,20))
print(np.cumsum(arr))  #累积和
```

# 比较函数

```python
print(np.greater([1,2,3],2))
print(np.less([1,2,3],2))
print(np.equal([1,2,3],2))
print(np.logical_and(1,0))
print(np.logical_or(1,0))
```

```python
print(np.any([1,0]))
print(np.all([1,1]))
```

```python
# 自定义条件
print(np.where(arr>10,arr,0))
```

```python
print(arr)
print(np.select([arr>10,arr<10],[10,5],default=1))
```

# 排序函数

```python
np.random.seed(1)
arr = np.random.randint(1,100,15)
print(arr)
print(np.sort(arr))
```

```python
print(np.unique(arr)) # 去重
arr1 = np.array([1,2])
arr2 = np.array([1,2])
print(np.concatenate((arr1,arr2))) #拼接
print(np.split(arr,3))# 等分
print(np.reshape(arr,[3,5]))
```
