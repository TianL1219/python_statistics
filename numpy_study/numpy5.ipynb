# numpy中的常用函数
## 基本数学函数

```python
import numpy as np
```

```python
#平方根
arr = np.array([1,2,3])
print(np.sqrt(arr))
```

```python
#e的指数
print(np.exp(1))
```

```python
#自然对数   lnX
print(np.log(2.718))
```

```python
#sin cos
print(np.sin(np.pi))
print(np.cos(np.pi))
print(np.tan(np.pi/2))
```

```python
#绝对值
print(np.abs(-1))
```

```python
#幂
print(np.power(2,3))
```

```python
#银行家四舍五入
print(np.round([4.5,5.4]))
```

```python
#向上取整  向下取整
arr = np.array([1.6,25.1,1.1])
print(np.ceil(arr))
print(np.floor(arr))
```

```python
#检测缺失值NaN
np.isnan(np.nan)
```
