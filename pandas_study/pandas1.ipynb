# series的创建

```python
import pandas as pd
import numpy as np
s = pd.Series([1,2,3,4],index = ['a','1','#','7'],name = 'aaa')
print(s)
```

```python
# 用字典创建
s = pd.Series({"a":100,"b":2,"c":3},name = "test")
print(s)
s1 = pd.Series(s,index=["a","c"])
print(s1)
```

```python
print(s.index)
print(s.values)
print(s.shape,s.ndim,s.size,s.dtype,s.name)
print(s.loc['a':'c'])
print(s.at['a'])
```

```python
#访问数据
print(s['a'])
print(s[s<3])
print(s.head())  #返回前五行
print(s.tail())  #后五行
```

```python
s = pd.Series([1,np.nan,None,2,3,4],index=['A','B','C','D','E','F'])
print(s.head())
print(s.tail(2))
```
